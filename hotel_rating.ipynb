```python
cols=['hotel_id', 'score_cleanliness',
       'score_comfort', 'score_facilities', 'score_location', 'score_staff',
       'score_value_for_money']
```

```python
x=df[cols]
y=df['score_overall']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
```

```python
base_models = [
    ('xgb', XGBRegressor(n_estimators=100, learning_rate=0.1, random_state=1)),
    ('decision_tree', DecisionTreeRegressor(max_depth=5, random_state=1)),
    ('svm', SVR(kernel='linear'))
]


# Meta model (final estimator)
meta_model = LinearRegression()
```

```python
stack=StackingRegressor(estimators=base_models,final_estimator=meta_model,cv=5)
```

```python
stack
```

```python
stack.fit(x_train,y_train)
```

```python
y_pred=stack.predict(x_test)
```

```python
r2=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
```

```python
print(r2)
print(mse)
```
